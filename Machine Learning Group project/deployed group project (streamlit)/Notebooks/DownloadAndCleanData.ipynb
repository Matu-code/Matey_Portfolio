```python
def remove_rows(df, column, labels):
    """
    Removes rows from a DataFrame based on specified labels in a given column.

    Args:
        df (pandas.DataFrame): The DataFrame from which rows will be removed.
        column (str): The column name in the DataFrame to filter rows.
        labels (list): A list of labels to identify the rows to be removed.

    Returns:
        pandas.DataFrame: The modified DataFrame with the specified rows removed.

    Example:
        df = pd.DataFrame({'Name': ['John', 'Jane', 'Alice', 'Bob'],
                           'Age': [25, 30, 28, 22],
                           'City': ['New York', 'London', 'Paris', 'Tokyo']})

        # Remove rows where City is 'London' or 'Paris'
        new_df = remove_rows(df, 'City', ['London', 'Paris'])
    """
    for label in labels:
        df = df.drop(df[df[column] == label].index)
    return df
```

```python
def sum_rows(df, column1, column2, value):
    """
    Calculates the sum of values in a specific column of a DataFrame based on a condition in another column.

    Args:
        df (pandas.DataFrame): The DataFrame from which the sum will be calculated.
        column1 (str): The column name to filter rows.
        column2 (str): The column name for which the sum will be calculated.
        value: The value in the column1 to match for filtering.

    Returns:
        int or float: The sum of values in column2 that satisfy the condition.

    Example:
        df = pd.DataFrame({'Category': ['A', 'B', 'A', 'B'],
                           'Value': [10, 20, 30, 40]})

        # Calculate the sum of 'Value' where 'Category' is 'A'
        total = sum_rows(df, 'Category', 'Value', 'A')
    """
    return df.loc[df[column1] == value, column2].sum()
```

```python
def melt_dataframe(df, id_column):
    """
    Restructures the columns of a DataFrame using the 'pd.melt' function.

    Args:
        df (pandas.DataFrame): The DataFrame to be restructured.
        id_column (str): The column name to use as the identifier column in the melted DataFrame.

    Returns:
        pandas.DataFrame: The restructured DataFrame obtained by melting the original DataFrame.

    Example:
        df = pd.DataFrame({'Country': ['USA', 'Canada', 'Mexico'],
                           '2018': [10, 20, 30],
                           '2019': [15, 25, 35],
                           '2020': [18, 28, 38]})

        # Restructure the DataFrame using 'Country' as the id_column
        melted_df = melt_dataframe(df, 'Country')
    """
    return pd.melt(df, id_vars=[id_column], value_vars = df.loc[:, df.columns != id_column])
```
