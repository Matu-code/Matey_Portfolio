```python
def dutch_to_english(text, batch=False): # Function to translate the data from Dutch to English
    if len(text)>1:
        try:
            if batch: 
                keywords=[]           
                for words in text:          
                    try:  
                        keywords.append(GoogleTranslator(source='auto', target='en').translate(words))
                    except:                    
                        num_word = num2words(float(words))
                        keywords.append(GoogleTranslator(source='auto', target='en').translate(num_word))
                return keywords
            return GoogleTranslator(source='nl', target='en').translate(text)
        except:
            return text
    else:
        return np.nan
```

```python
# Translate the title column
df_content['title'] = [dutch_to_english(x) for x in tqdm(df_content['title'])]
```

```python
# Translate the summary column
df_content['summary'] = [dutch_to_english(x) for x in tqdm(df_content['summary'])] 
```

```python
# Translate the keywords column
df_content['keywords'] = [dutch_to_english(x, batch=True) for x in tqdm(df_content['keywords'])]
```

```python
df_content["date"] = pd.to_datetime(df_content["date"], infer_datetime_format=True)
```

```python
# Remove the milleseconds
df_content["start"] = df_content["start"].str[:-3]
df_content["end"] = df_content["end"].str[:-3]

df_content["date_time_start"] = pd.to_datetime(df_content["date"].astype(str) + ' ' + df_content["start"].astype(str), infer_datetime_format=True)
df_content["date_time_end"] = pd.to_datetime(df_content["date"].astype(str) + ' ' + df_content["end"].astype(str), infer_datetime_format=True)
```

```python
df_content["content_id"] = df_content["id"].apply(lambda x: x[14:28])
df_content["fragment"] = df_content["id"].apply(lambda x: x[27:28])
```
